```python
import matplotlib.pyplot as plt

# Define data
data = [
    {'name': 'Rainne', 'rating': 8},
    {'name': 'Buddy', 'rating': 7},
    {'name': 'Paddy', 'rating': 3},
    {'name': 'Sticky', 'rating': 9},
    {'name': 'Midnight', 'rating': 5},
    {'name': 'Leo', 'rating': 6}
]

# Extract names and ratings
names = [d['name'] for d in data]
ratings = [d['rating'] for d in data]

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the line
ax.plot(names, ratings, marker='o', linestyle='-', color='pink', linewidth=2)

# Add labels and title
ax.set_xlabel('Names')
ax.set_ylabel('Rating')
ax.set_title('Ratings of Different Names')

# Add points (optional since they're already in the line plot)
ax.scatter(names, ratings, color='purple', s=50, zorder=5)

# Show the plot
plt.xticks(rotation=45)
plt.show()
```
